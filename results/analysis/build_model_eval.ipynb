```python
import os

model_dir='..\\models\\hyperparam'
output_file='..\\..\\runner\\config\\hyperparam_models.txt'

search = {
    "algos": ["iql", "qmix"],
    "map_names": ["confidentiality_medium"],
    "batch_size": ["256", "128"],
    "buffer_size": ["5000", "10000"],
    "lr": ["0.005", "0.01"],
    "td_lambda": ["0.0", "0.6"]
}


def select_models(models, search):
    output_models = {}

    for model in models:
        config = []
        
        for algo in search["algos"]:
            if algo in model: 
                #for map_name in search["map_names"]:
                #    if map_name in model:
                for field, values in search.items():
                    for value in values:
                        if f"{field}={value}" in model:
                            if model in output_models:
                                output_models[model].append(f"{field}={value}")
                            else:
                                output_models[model] = [algo, f"{field}={value}"]
    print(output_models)
    return output_models

def make_model_list(root_dir):
    models = []

    for root, dirs, files in os.walk(root_dir):
        if root == root_dir:
            for model in dirs:
                models.append(os.path.join(os.path.abspath(root), model))

    return models
        
models =  make_model_list(model_dir)
output_models = select_models(models, search)

#with open(output_file, 'w') as f:
#    f.write("\n".join(map(str, output_models)))

with open(output_file, 'w') as f:
    for key, value in output_models.items():
        line = ','.join(value) + ',' + key
        f.write(line + '\n')
```
